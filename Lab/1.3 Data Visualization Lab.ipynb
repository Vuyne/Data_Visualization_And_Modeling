```python
import pandas as pd

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.tail()
```

```python
df['bpm'].plot(kind = 'hist',
               title = 'Distribution Of BPM Spotify Hit Song in 2023',
               xlabel = 'Beats Per Minites',
               ylabel = 'Number of Songs',
               bins=20
              )
```

```python
df['bpm'].describe()
```

The distribution of BPM is in bell-shaped, most songs falling between 120 and 140 BPM. Therefore, a guaranteed hit song have a tempo between 120 and 140 BPM.

```python
key_counts = df['key'].value_counts()
key_counts
```

```python
key_counts.plot(kind = 'bar',
               ylabel = 'count',
               title = 'Distribution Of Key Of Spotify Hit Song in 2023' )
```

C# is the most common key, appearing in approximately 120 songs, followed by G and G# . The least common key is D#, with only about 33 songs. Overall, the keys are relatively spread across all 11 values, but there is a clear preference for C#, making it the ideal key for a guaranteed hit song.

```python
mode_counts = df['mode'].value_counts()
mode_counts.plot(kind = 'bar',
               ylabel = 'count',
               title = 'Distribution Of Mode Spotify Hit Song in 2023' )
```

Major mode is significantly more common, with approximately 550 songs, compared to Minor with around 400 songs. This suggests that hit songs tend to favor a brighter, more upbeat sound.

```python
dance_counts = df['danceability_%'].value_counts()
dance_counts
```

```python
df['danceability_%'].plot(kind = 'hist',
                          xlabel = 'Danceability(%)',
               title = 'Distribution Of Danceability Of Spotify Hit Song in 2023' )
```

```python
df['danceability_%'].describe()
```

The distribution is approximately bell-shaped, with a peak between 69% and 78% (frequency ~200), indicating that most hit songs have moderately high danceability. Values below 40% and above 90% are relatively rare.

```python
df['energy_%'].plot(kind = 'hist',
                    xlabel = 'Energy(%)',
                    bins=20,
                    title = 'Distribution Of Energy Of Spotify Hit Song in 2023' )
```

```python
df['energy_%'].describe()
```

The distribution is approximately bell-shaped with a slight left skew, peaking between 60% and 65% (frequency ~113). Most songs are concentrated in the 50% to 80% range, while very few songs fall below 20%. Therefore, a guaranteed hit song should have an energy score between 66% and 77%."

```python
df['speechiness_%'].plot(kind = 'hist',
                    xlabel = 'speechiness(%)',
                         bins =20,
                    title = 'Distribution Of Speechiness Of Spotify Hit Song in 2023' )
```

The distribution is strongly right-skewed, with the vast majority of songs having very low speechiness. The peak is at 5% (frequency ~460). Very few songs exceed 40%, and almost none go beyond 50%. A hit song should have a speechiness score between 3% and 8%."

```python
df['acousticness_%'].plot(kind = 'hist',
                    xlabel = 'acousticness(%)',
                          bins=20,
                    title = 'Distribution Of Acousticness Of Spotify Hit Song in 2023' )
```
